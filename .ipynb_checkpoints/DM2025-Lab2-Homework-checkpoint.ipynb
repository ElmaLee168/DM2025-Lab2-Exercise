```python
#下載資料
import json
import pandas as pd

# final_posts.json 是 complex nested structure
with open("data/final_posts.json", "r", encoding="utf-8") as f:
    raw_data = json.load(f)

# 每一筆資料是 raw_data[i]["root"]["_source"]["post"]
posts = []
for item in raw_data:
    post = item["root"]["_source"]["post"]
    posts.append({
        "id": post["post_id"],
        "text": post["text"]
    })

df_posts = pd.DataFrame(posts)

print(df_posts.head())
print(df_posts.columns)


```

```python
# 合併資料

# 讀 CSV
df_id = pd.read_csv("data/data_identification.csv")      # ['id', 'split']
df_emotion = pd.read_csv("data/emotion.csv")             # ['id', 'emotion']

# 合併成完整資料
df = df_id.merge(df_posts, on="id", how="left")
df = df.merge(df_emotion, on="id", how="left")

# 切成 train/test
train_df = df[df["split"] == "train"].reset_index(drop=True)
test_df = df[df["split"] == "test"].reset_index(drop=True)

print("Train:", len(train_df))
print("Test:", len(test_df))

train_df.head()
```

## 3. Model Implementation Steps

```python
#先將情緒編號才能訓練
# 建立 label 編號（0~num_classes-1）
unique_emotions = train_df["emotion"].unique().tolist()
label2id = {label: idx for idx, label in enumerate(unique_emotions)}
id2label = {idx: label for label, idx in label2id.items()}

train_df["label"] = train_df["emotion"].map(label2id)
```

```python
#Tokenizer + Dataset
from transformers import AutoTokenizer
import torch
from torch.utils.data import Dataset

model_name = "distilbert-base-uncased"   # 較快又準

tokenizer = AutoTokenizer.from_pretrained(model_name)

class EmotionDataset(Dataset):
    def __init__(self, df, tokenizer, is_train=True):
        self.texts = df["text"].tolist()
        self.is_train = is_train
        if is_train:
            self.labels = df["label"].tolist()
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]

        enc = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=64,
            return_tensors="pt"
        )

        item = {key: val.squeeze(0) for key, val in enc.items()}

        if self.is_train:
            item["labels"] = torch.tensor(self.labels[idx])
        return item

#train_dataset = EmotionDataset(train_df, tokenizer, is_train=True)
#test_dataset = EmotionDataset(test_df, tokenizer, is_train=False)

# 先抽樣一份小訓練集
df_small = train_df.sample(n=2000, random_state=42).reset_index(drop=True)

train_dataset = EmotionDataset(df_small, tokenizer, is_train=True)
test_dataset = EmotionDataset(test_df, tokenizer, is_train=False)


#檢查
sample_text = train_df["text"].iloc[0]
enc = tokenizer(
    sample_text,
    truncation=True,
    padding="max_length",
    max_length=50,
    return_tensors="pt"
)

print("原始句子：")
print(sample_text)

print("\nTokens：")
print(tokenizer.convert_ids_to_tokens(enc["input_ids"][0]))

print("\ninput_ids shape:", enc["input_ids"].shape)
print("attention_mask shape:", enc["attention_mask"].shape)
```

```python
# 建立 DataLoader
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)

batch = next(iter(train_loader))

for k, v in batch.items():
    print(k, v.shape)
```

```python
# 建立模型
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(unique_emotions)
)

# 訓練設定
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="bert_output",
    per_device_train_batch_size=4,
    num_train_epochs=1,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_steps=50,
    
    eval_strategy="no",      # ← 正確參數名稱
    save_strategy="epoch",   
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)
trainer.train()
```

```python
# 檢查模型
model.eval()
with torch.no_grad():
    outputs = model(
        input_ids=batch["input_ids"],
        attention_mask=batch["attention_mask"],
        labels=batch["labels"]
    )

print(outputs)
```

```python
# 快速檢查 single prediction 是否正常
text = test_df["text"].iloc[0]
enc = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

model.eval()
with torch.no_grad():
    out = model(**enc)

pred = out.logits.argmax(dim=1).item()
print("預測類別 id:", pred)
print("預測情緒:", id2label[pred])
```

```python
# 產生kaggle submission file
import pandas as pd

# 預測
pred_output = trainer.predict(test_dataset)
logits = pred_output.predictions
# 取最大值（分類結果）
pred_ids = logits.argmax(axis=1)

# 轉回文字類別 (string labels)
pred_emotions = [id2label[i] for i in pred_ids]

# 準備建立csv檔案
submission = pd.DataFrame({
    "id": test_df["id"],
    "emotion": pred_emotions          # 文字標籤
})

submission.to_csv("submission.csv", index=False, encoding="utf-8")
print("submission.csv 已輸出！")

submission.head()
```
